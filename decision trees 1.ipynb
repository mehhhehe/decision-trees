```python
class Cart:
    def __init__(self, X, y, max_depth = 10):
        
        self.X = X
        self.y = y
        self.features = list(self.X.columns)
        self.target = list(self.y.columns)[0]
        self.metadata = list()
        self.Tree = list()
        self.max_depth = max_depth
        
    def gini_ind_feat(self, feature, y_set):
        # !!!! fix this - gini index should be based on the no. of samples in y
        X_set = pd.DataFrame(data=[self.X.loc[i] for i in y_set.index], index=y_set.index)
        values = {val: {i:0 for i in y_set[self.target].unique()} for val in X_set[feature].unique()}
        total_samples = len(y_set)
        # print(total_samples)
        for index in list(y_set.index):
            for val in (y_set[self.target]).unique():
                a = X_set.loc[index][feature]
                if y_set.loc[index][self.target] == val:
                    values[a][val] += 1

        
        gini_indices = {}
        for val in values.keys():
            
            temp = {i:0 for i in y_test[self.target].unique()}
            
            for tar in values[val].keys():
                temp[tar] = values[val][tar]
            
            tot = sum(temp.values())

            summ = 1
            # print(temp)
            for j in temp.values():
                x = ((j/tot)**2)
                # print(x)
                summ -= x
            gini_indices[val] = (tot/total_samples)*summ    
        # print(gini_indices)
        # print(f'Sum: {sum(gini_indices.values())}')
        return sum(gini_indices.values())

    def gini_ind_set(self, y_set):
        # print(len(y_set))
        vals = y_set[self.target].unique()
        freq = {i: list(y_set[self.target]).count(i) for i in vals}
        # print(freq)
        total = sum(freq.values())
        gini = 1
        for i in freq.values():
            gini -= (i/total)**2
            
        return gini

    def delta_gini(self, feature, y_set):
        # print(f'Gini index of set: {self.gini_ind_set(y_set)} - Gini index of feature: {self.gini_ind_feat(feature, y_set)}')
        return self.gini_ind_set(y_set) - self.gini_ind_feat(feature, y_set)

    def split(self, y_set, count = 0, rule = dict()):
        # print(f'Count: {count}')
        if len(y_set) < 1 or self.gini_ind_set(y_set) == 0 or count > self.max_depth:
            if len(y_set) > 0:
                targ = list(dict(sorted(dict(y_set[self.target]).items(), key=lambda x:x[1])[::-1]).items())[0]
                return targ[1]
            return None    
            
        rank = {feature: 0 for feature in self.features}
        for feat in self.features:
            rank[feat] = self.delta_gini(feat, y_set)
        ranks = dict(sorted(rank.items(), key=lambda x:x[1])[::-1])
        # print(rank)
        feature_to_split = list(ranks.keys())[0]
        
        values = set(self.X[feature_to_split])
        # print(values)
        rows = {val:[ind for ind in y_set.index if self.X.loc[ind][feature_to_split] == val] for val in values}
        # print(rows)
        temp = {val:{i: 0 for i in set(y_set[self.target])} for val in values}

        for ind in y_set.index:
            for val in values:
                print(temp[val])
                # print(self.X.loc[ind][feature_to_split])
                # print(y_set.loc[ind][self.target])
                if val == self.X.loc[ind][feature_to_split]:
                    temp[val][y_set.loc[ind][self.target]] += 1
                    break
                
        self.metadata.append(f'Splitting on {feature_to_split}, level = {count}, set = {temp}\n\n')             
        i = count + 1
        for val in values:
            new_set = pd.DataFrame(data = [y_set.loc[i] for i in rows[val]], index = rows[val])
            rule[feature_to_split] = {f'{val}': self.split(new_set, count=i, rule = rule)}
            self.Tree.append(rule)
                
            
            
        
            
```
